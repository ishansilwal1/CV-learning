# tensorflow basics


```python
import tensorflow as tf

tensor_zero_d = tf.constant(4)

print(tensor_zero_d)
```

```python
tensor_one_d = tf.constant([2,1,2])


print(tensor_one_d)
```

```python
tensor_two_d = tf.constant([
    [ 1,2,0],
    [1,2,4],
    [0,0,0]
])

print(tensor_two_d)
```

```python
tensor_3_d = tf.constant([
    [[1, 2, 0], [1, 2, 5]],  # First 2x3 matrix
    [[1, 2, 0], [1, 2, 5]]   # Second 2x3 matrix
])

print(tensor_3_d)
```
