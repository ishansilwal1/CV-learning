6. Create a function that resizes an image to 40% of it's original size 

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 

img = cv.imread('D:\CV practice\learning\image2.jpg')
def rescaleFrame(frame,scale=0.75):
    width = int(frame.shape[1]*scale)
    height = int(frame.shape[0]*scale)
    dimensions = (width,height)
    return cv.resize(frame,dimensions,interpolation=cv.INTER_AREA )

resized_image = rescaleFrame(img,0.4)
plt.imshow(cv.cvtColor(resized_image,cv.COLOR_BGR2RGB))
```

9. Rotate an image by 45 degrees without cropping ( Using rotation matrix + bounding)

```python
import cv2 as cv 
import matplotlib.pyplot as plt 
img = cv.imread("D:/CV practice/learning/image2.jpg")
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

def rotation (img,angle,rotpoint=None):
    (height,width) = img.shape[:2]
    if rotpoint is None:
        rotpoint = (width//2,height//3)
    rotmat = cv.getRotationMatrix2D(rotpoint,angle,1.0)
    dimensions = (width,height)
    return cv.warpAffine(img,rotmat,dimensions)

rotated = rotation(img,45)

plt.imshow(rotated)
```

10. Read a video file and count how many frames it contains 

```python
import cv2 as cv 
import matplotlib.pyplot as plt 
import numpy as np 
from IPython.display import HTML

# Path to your video
video_path ="D:/CV practice/learning/race_car.mp"

# Open the video
cap = cv.VideoCapture(video_path)

if not cap.isOpened():
    print("Error opening video file!")
else:
    # Total number of frames (fast)
    total_frames = int(cap.get(cv.CAP_PROP_FRAME_COUNT))
    print("Total Frames:", total_frames)

cap.release()


```

10. Write a script to draw a grid on an image (like dashboard )

```python
import cv2 as cv 
import numpy as np 

img = np.zeros((512,512,3),dtype="uint8")

h,w,c = img.shape
row = 3 
col = 4

row_h = row//h
col_h = col//w

cv.line(img, (row_h, 0), (col_h, h), (0, 255, 0), 2)
cv.imshow(img)


cv.waitKey(0)
```
