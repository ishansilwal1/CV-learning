# Activation Functions 

It is a mathematical function applied to the outcome of a neuron. It introduces non-linearity, enabling the model to learn and represent complex data patterns. 

It decides whether a neurn should be activated based on the weighted sum of inputs and a bias term . 


# WHy important ? 

They allow neural network to form curved Descision boundaries, making them capable of handling comlpex patterns . as they introduce non-linearity. 

Without activation function even a deep neural network will behave like a simple linear regression model  .
