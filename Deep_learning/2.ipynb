# Perceptron Loss Function 

Perception Loss function is the mathematical tool to find out how poorly a perception is classifying the training data. 


Perception Loss function focuses on misclassified points to find the error

# HOW IT WORKS ( maths)

A perception makes a prediction by calculating the weighted sum 

f(x) = weights.inputs + bias 

The formula for perception loss is 

:
L(y,f(x))=max (0,-y.f(x))
