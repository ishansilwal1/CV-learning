# Convolutional Neural Network 

It is  a neural network designed to extract features from grid like matrix datasets. 
It is useful for visual datasets such as images or videos. 

It contains multiple layers like input, covolutional layer , pooling layer and fully connected layers. 

# Convolutional Operation


Layers before find simpler features like edges , 
layers after that find some complex pattern and by adding them all a complete image is identified. 

<h3> Convolutional Layer </h3>

It is a layer designed to process grid like data like images and videos. 
It is the primary building block of a CNN  and it automatically extracts features like edges, textures and shapes from input by sliding filter(Kernel) across it . 

Filter or Kernel is a small matrix that slides over the image acting as the pattern detector. The filer value is similar to let's say weights in ANN 

<h3> Padding and Strides </h3>          

Padding = process of adding extra pixels, typically zeros around the border of the input image or feature map. . 

Padding and strides are important concepts that determines how the convolutional operation is applied to an input, affecting the output size and how features are extracted. 


Strides = number of pizels by which the filter (kernel) moves or slides across the input image. 


 




stride increase == feature map size decrease 

# Pooling Layer in CNN 

It is used to reduce the domentions of the image feature maps while retaining the most important information

<h3> Why Important </h3>

1. Dimensionality Reduction: Pooling layer reduces the spatial size of the feature maps, which decreases the number of parameters and coputations in the network. It makes the model faster and more efficient 

2. It helps network become invariant to small translations or distributions in the input image.  example : even if an object in an image is slightl
