# WHat is loss function ? 

It is a method to evaluate how good the algorithm/model is modelling our dataset 

 It is a mathematical way to measure how good or bad a model's prediction are compared to the actual results.   It simply gives us a single number that tells us how far off the prediction are .



  Smaller Number  == Better Predictions
  

They are important as they: 

1. Guide Model Training : 
                        During training algorithm like gradient descent use the loss function to adjust the model's parameters and try to reduce the error and improve model's prediction. 

2. Measure MOdel Performance : 
                                Can be used in model evaluation by finding difference between predicted values and actual values. 
3. Affect model learning behaviour: 
                                    Different loss functions can make model learn in differnt ways based on the mistakes they make

To measure the errors and improve the model performance
