# unet Segmentation 

 Unet is a kind of neural network used for image segmentation. 
It means dividing an image into different parts or segments based on what is in the image.

The Architecture is symmetric and has three key parts. 







<h3> 1.  Contracting Path (Encoder) </h3>

 In this part small filters of (3x3pixels) is used to scan images and features, activation function relu is applied to add non-linearity and help the model to learn better and max pooling is used (2x2 filter) to shrink the image size while keeping the most important featuers. 

<h3> 2. Bootle Neck </h3>

It is the middle of the "U" shape where the most compressed and abstract information is stored . It links the encoder and decoder. 


# Main U-Net Architecture

<h3> 1. Contracting Path (Encoder)</h3>

<h4>What is it?</h4>
The encoder is the left side of the U shape. Its like a funnel that takes the input image and gradually makes it smaller, but richer in information.

<h3>How does it work?</h3>

The image passes through several blocks.
Each block uses small filters (3x3) to scan for patterns like edges, corners, and textures. This is called a convolution.
After each set of convolutions, a ReLU activation function is applied to help the network learn complex patterns.
Then, max pooling (2x2) is used to shrink the image size by half, keeping only the most important information.
As the image gets smaller, the number of channels (feature maps) increases, so the network can learn more features.

<h4>Why do this?</h4>
Shrinking the image helps the network focus on the most important features and ignore unnecessary details, while increasing channels allows it to learn more complex information.

<h3>2. Bottleneck</h3


<h4> What is it?</h4>

The bottleneck is the bottom of the U. Here, the image is at its smallest size but has the most channels (features).

<h3>How does it work?</h3>

The network applies more convolutions to this compressed version of the image.
This part captures the most abstract and high-level features of the image.

<h4>Why do this?</h4>
The bottleneck acts as a bridge between the encoder and decoder, summarizing all the important information from the input image.

<h3>3. Expanding Path (Decoder/Upsampling) </h3>

<h4> What is it?</h4>
The decoder is the right side of the U. It takes the compressed information and reconstructs the image back to its original size, but now each pixel is labeled with its class.

<h3>How does it work? </h3>

The decoder uses upsampling (transposed convolutions) to make the image bigger step by step.
At each step, it also brings back information from the matching layer in the encoder (these are called skip connections). This helps recover fine details lost during shrinking.
After concatenating with the encoders features, more convolutions are applied to refine the information.
As the image gets bigger, the number of channels decreases, focusing the information into a detailed segmentation map.


<h3>Why do this ? </h3>
Upsampling allows the network to create a detailed map of the image, showing exactly where each object or region is, using both the big-picture understanding and the fine details.
