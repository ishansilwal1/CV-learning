# Cost function in Linear Regression . 

In linear regression , cost function measures how far the predicted values are from the actual values. 
it helps to identify the error and reduce them to find the best fit line. 


In linear regression The most common  used cost function is Mean Squared error . 

<h3> Mean Squared Error </h3>
It calculates the average of squared differences between the actual and the predicted values. 

To minimize the cost we use gradient descent which takes MSe to its lowest values. and ensure the line fits the data as accurately as possible. 


# Gradient Descent for Linear Regression 


It is an optimization technique used to train a linear regression model by minimizing the prediction error. 

It works by starting with random model parameter and repeatedly adjusting them to reduce the difference between actual and predicted values. 


<h3> Example work </h3>

It starts with random values for slope and intercept . 
Calculates the error between predicted and actual values. 
find how much each parameter contributes to the error ( which is called gradient )
Change the parameter to reduce the error and update them to the value when the error is as small as possible. 



<h2> Mean Square Error (MSE )</h2>
It is an evaluation metrics that calculates the average of squared differences between the actual and predicted values for all the data points . 

It is a way to quantify the accuracy of a model's predictio . 
It is sensitive to outliers .

<h2> Mean Absolute Error </h2>

It measures the average of the absolute difference between the predicted and the actual values. 

<h4> Low MAE == Better Model Performance </h4>

Not sensitive to outliers.
