# Encoding Categorical Values 



1. nominal categorical data 
     has no relation between categories 
     example : state ( each one of them is equal ) or branch in engineering 


2. Ordinal categorical data 
     has relation between categories . 
     example : marks in high school( One is better or worst than other )
     or let's say ratings 




<h3> How to encode categorical data? </h3>

The categorical data is in string format mostly and ML algorithms expect numbers . so they are converted to numerical values using encoding 


Example/Most Popular ones  : Ordinal encoding , Label encoding and One hot encoding. 




<h3> Label Encoding  </h3>
if target variable is categorical (classification Problem) let's say output is categorical then label encoder is used . It works the same as ordinal encoding but is used in output labels. 







<h3> Ordinal Encoding </h3>
It works in input variables/column . 

Used in tree based model like decision tree or XGBOost





```python
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Green', 'Blue', 'Red']

le = LabelEncoder()
encoded_data = le.fit_transform(data)
print(f"Encoded Data: {encoded_data}")
```

<h3> One Hot Encoding </h1>

It converts categorical column to binary columns. It prevents false ordering but can be the reason for high dimensionality for too many unique values.



<h4> When it is used ? </h4>

Used when dealing with linear modesls , logistic regression and neural networks. 


```python
# Example implemented .
import pandas as pd
data = ['Red', 'Blue', 'Green', 'Red']

df = pd.DataFrame(data, columns=['Color'])
one_hot_encoded = pd.get_dummies(df['Color'])

print(one_hot_encoded)
```

In the above example each unique category 'red', 'blue' , 'green' is transformed to a seperate binary column . 1 /True = presence of category and 0/False means absence.
