```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 


## drawing circles lines rectangles , adding lines and texts

img = cv.imread('image2.jpg')
cv.imshow('original',img)


plt.imshow(img[:,:,::-1])
cv.waitKey(0)
```

```python
img = cv.line(img,(150,250),(255,350),(255,0,0),5)
plt.imshow(img[:,:,::-1])
```

```python
image = cv.rectangle(img,(300,200),(400,300),(0,255,0),3)
plt.imshow(img[:,:,::-1])
```

```python
cv.rectangle(img,(320,190),(500,370),(0,0,255),4)
plt.imshow(img[:,:,::-1])
```

```python
img = cv.imread('image2.jpg')
# plt.imshow(img[:,:,::-1])
cv.putText(img,'Check',(300,100),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)

plt.imshow(img[:,:,::-1])
```

```python
cv.putText(img,'Batman',(350,370),cv.FONT_HERSHEY_COMPLEX,-1,(255,0,0),3)

plt.imshow(img[:,:,::-1])
```

```python
img = cv.imread('image2.jpg')

cv.circle(img,(400,300),100,(0,255,255),-2)
plt.imshow(img[:,:,::-1])
```
